# Seaborn: scatterplot

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')

diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds.cut.unique()
```

```python
diamonds = diamonds[
            diamonds.cut.isin(['Premium', 'Good']) & 
            diamonds.color.isin(['D', 'F', 'J'])
            ].sample(n=100, random_state=22)

diamonds.cut.cat.remove_unused_categories(inplace = True)
diamonds.color.cat.remove_unused_categories(inplace = True)

diamonds.shape
```

```python
diamonds.cut
```

```python
diamonds.cut.value_counts()
```

```python
diamonds.color
```

```python
diamonds.color.value_counts()
```

## Semantic Variables

```python
sns.set_style('darkgrid')
```

### hue

```python
plt.figure(figsize = (8, 8))
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds
               );
```

```python
sns.scatterplot(x='carat', y='price', hue='color',
                data=diamonds
                ,palette=['green','#55CCCC', 'gray']
               );   # how to choose colors? what options are available?  will check in slides 
# https://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                data=diamonds,
                palette = ['purple', '#55CCCC'],
                hue_order=['Good', 'Premium']
               );
```

#### Continuous Variable

```python
plt.figure(figsize = (8, 6))
sns.scatterplot(x='carat', y='price', hue='depth', 
                data=diamonds, s = 100);  # how to increase size of dots? 
```

### size
