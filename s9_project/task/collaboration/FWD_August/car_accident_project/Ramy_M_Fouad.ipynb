```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv('df.csv')
```

```python
plt.figure(figsize=[9,6])
sns.set_style('darkgrid')
sns.barplot(y='Distance(mi)',x='Severity',data=df)

sns.despine()
#plt.xticks([1,2,3,4])
plt.title('Relation Between Distance And Severity Of Accident',fontsize=18,color="green")
plt.ylabel('Distance by mi ',rotation=0 ,color= "gray",fontsize=14,labelpad=50)
plt.xlabel("Sevirity Grade",color="gray",fontsize=14)
plt.show();
```

```python
df_vis=df[df['Visibility(mi)']<=10]
```

```python
sns.set_style("white")
sns.kdeplot(x=df_vis['Visibility(mi)'],hue=df_vis["Severity"])
sns.despine()
plt.ylabel("")
plt.title("KDE of visibility in accidents ",fontsize=20,pad=30,color='green')
plt.xlabel("Visibility",fontsize=14)
plt.xlim(0,)
plt.show();
```

```python
plt.figure(figsize=[6,5])
sns.lineplot(x="Visibility(mi)",y='Temperature(F)',data=df_vis)
sns.despine()
plt.title("Relation Between Visibility And Tempreture",fontsize=20,pad=30,color='green')
plt.ylabel("Temp",rotation=0,fontsize=14,labelpad=30)
plt.xlabel("Visibility",fontsize=14)
plt.xlim(0)
plt.ylim(0)
plt.show();
```
